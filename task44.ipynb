```python
import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
pd.get_dummies(lst)
# как работает с get_dummies , который надо заменить.
```

```python
lst_human = ['']*lst.__len__()
lst_robot = ['']*lst.__len__()
for i in range(lst.__len__()):
    if lst[i] == 'human':
        lst_human[i] = 'ТруЪ'
        lst_robot[i] = 'False'
    if lst[i] == 'robot':
        lst_human[i] = 'Нелюдь'
        lst_robot[i] = 'True'
data = pd.DataFrame({'human':lst_human, 'robot':lst_robot})
data
```
